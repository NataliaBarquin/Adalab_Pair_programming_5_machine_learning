```python
import pandas as pd
import numpy as np
import datetime as dt

import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.multivariate.manova import MANOVA
from sklearn.preprocessing import StandardScaler

import warnings
warnings.filterwarnings('ignore')

pd.options.display.max_columns = None 
```

```python
df = pd.read_pickle('data/supermarket_estand.pkl')
df.head(2)
```

```python
#Realizamos el test de ANOVA
#Nuestra variable respuesta el 'quantity' (la cantidad de productos en cada ticket)

lm = ols('total_boxcox ~ city  +customer_type + gender + product_line + unit_price + quantity + payment + rating + month + shift', data=df).fit()
sm.stats.anova_lm(lm)
```

```python
lm.summary()
```
